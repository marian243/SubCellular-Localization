```python
models = []
val_categories = []
for category in categories:
    X_sample = X_train
    y_sample = y_train[category]

    sampler = RandomUnderSampler(sampling_strategy=1)
    
    X_sample, y_sample = sampler.fit_resample(X_sample, y_sample)

    if category == 'cytoplasm' or category == 'cell membrane':
        size_up = 30
        size_bottom = y_sample.value_counts()[1]
        size = math.ceil((size_bottom/100)*size_up)
        
        c = 0
        index = []
        for i in range(y_sample.shape[0]):
            if y_sample[i] == 1 and c <= size:
                c += 1
                index.append(i)
            
        X_more = X_sample.iloc[index,:]
        y_more = y_sample.iloc[index]   
        
        print(y_sample.value_counts())
        
        X_sample = X_sample.append(X_more, ignore_index=True)
        y_sample = y_sample.append(y_more, ignore_index=True)
        
    print(y_sample.value_counts())
    
#     print(X_sample)
#     print(y_sample)
    
    classifier.fit(X_sample, y_sample)
    models.append(classifier)
    y_pred_proba = classifier.predict_proba(X_val)
    
    y_pred = []
    y_prob = []
    for i in y_pred_proba:
#         print(i[1])
        y_prob.append(i[1])
        if i[1] >= 0.6:
            y_pred.append(1)
        else:
            y_pred.append(0)
    val_categories.append(category+'_pred')
    df_prob[category+'_pred'] = y_prob
    df_val[category+'_pred'] = y_pred
```
